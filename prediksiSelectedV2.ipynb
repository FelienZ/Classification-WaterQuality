```python
import pandas as pd
import pickle

# Load model
with open("water_quality_rf_modelV3.pkl", "rb") as f:
    model = pickle.load(f)

print("=== INPUT DATA AIR ===\n")

# Inputan
pH = float(input("Masukkan pH (contoh 7.1): "))

print("\nColor:")
print("0 = Colorless")
print("1 = Faint Yellow")
print("2 = Light Yellow")
print("3 = Near Colorless")
print("4 = Yellow")
color = int(input("Masukkan kode Color: "))

print("\nTurbidity:")
print("0 = Jernih")
print("1 = Agak keruh")
print("2 = Sangat keruh")
turbidity = int(input("Masukkan kode Turbidity: "))

print("\nOdor:")
print("0 = Bau")
print("1 = Tidak Bau")
odor = int(input("Masukkan kode Odor: "))

print("\nSource:")
print("0 = Aquifer")
print("1 = Ground")
print("2 = Lake")
print("3 = Reservoir")
print("4 = River")
print("5 = Spring")
print("6 = Stream")
print("7 = Well")
source = int(input("Masukkan kode Source: "))

violations = []

if not (6.5 <= pH <= 8.5):
    violations.append(f"pH tidak aman: {pH} (Standar WHO: 6.5 - 8.5)")
if turbidity > 5:
    violations.append(f"Turbidity terlalu tinggi: {turbidity} (Standar WHO: ≤ 5 NTU)")
# if tds > 500:
#     violations.append(f"TDS terlalu tinggi: {tds} (Standar WHO: ≤ 500 mg/L)")
# if iron > 0.3:
#     violations.append(f"Kadar Iron terlalu tinggi: {iron} mg/L (Standar WHO: ≤ 0.3 mg/L)")
# if manganese > 0.1:
#     violations.append(f"Kadar Manganese terlalu tinggi: {manganese} mg/L (Standar WHO: ≤ 0.1 mg/L)")

if len(violations) > 0:
    print("\nHASIL CEK WHO: AIR TIDAK LAYAK MINUM")
    print("Alasan:")
    for v in violations:
        print(" -", v)
    raise SystemExit ("Proses dihentikan karena melanggar standar WHO.") # Tidak melanjutkan ke model

# DataFrame input
user_input = pd.DataFrame([{
    "pH": pH,
    "Color": color,
    "Turbidity": turbidity,
    "Odor": odor,
    "Source": source
}])

# Menambahkan fitur yang tidak diinput user dengan imputation median
extra_cols = ["Total Dissolved Solids", "Manganese", "Iron"]
for col in extra_cols:
    user_input[col] = None   # None nanti diimpute oleh model

prediction = model.predict(user_input)[0]
proba = model.predict_proba(user_input)[0]

print("\n=== HASIL PREDIKSI MODEL ===")
if prediction == 0:
    print("Layak Minum")
else:
    print("Tidak Layak Minum")

print(f"\nProbabilitas Tidak Layak: {proba[0]:.3f}")
print(f"Probabilitas Layak:       {proba[1]:.3f}")
```
